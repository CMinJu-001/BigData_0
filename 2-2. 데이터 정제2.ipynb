# 1. 데이터 불러오기

```python
import pandas as pd
```

```python
평일_추가전 = pd.read_csv('C:/python/2-2. 원본데이터/외곽선_평일_추가전.csv', encoding='EUC-KR', index_col = 0)
평일_추가전
```

```python
주말_추가전 = pd.read_csv('C:/python/2-2. 원본데이터/외곽선_주말_추가전.csv', encoding='EUC-KR', index_col = 0)
주말_추가전
```

```python
종별교통량 = pd.read_csv('C:/python/2-2. 원본데이터/외곽선_2019_교통량.csv', encoding='EUC-KR')
종별교통량 = 종별교통량.iloc[:,:-2]
종별교통량
```

```python
차로 = pd.read_csv('C:/python/2-2. 원본데이터/VDS_ID별 차로개수 및 지점이정.csv',encoding="EUC-KR",index_col=0)
수도권_차로 = 차로[차로['VDS_ID'].str.contains("1000VD")] # 수도권제1순환고속도로 1000
수도권_차로
```

# 2. 변수 "집계시" int -> str 변환

종별 교통량 변수 "집계시" : object (합계 행 포함)

평일_추가전, 주말_추가전 변수 "집계시" : int

데이터 프레임 합치기 실행 시 오류 발생 → 평일_추가전과 주말_추가전의 변수 "집계시" object로 변환 실시

```python
종별교통량.dtypes
```

```python
평일_추가전['집계시']=평일_추가전['집계시'].astype(str)
평일_추가전.dtypes
```

```python
주말_추가전['집계시']=주말_추가전['집계시'].astype(str)
주말_추가전.dtypes
```

# 3. 변수 "콘존명" 에 '-' 행 삭제

수도권제1순환고속도로의 민자구간인 퇴계원IC-일산IC는 분석에서 제외하고 한국도로공사에서 운영하고 있는 나머지 구간만 추출

```python
평일_추가전 = 평일_추가전[평일_추가전['콘존명']!="-"].reset_index(drop=True)
평일_추가전
```

```python
주말_추가전 = 주말_추가전[주말_추가전['콘존명']!="-"].reset_index(drop=True)
주말_추가전
```

# 4. 변수 추가

## 4-1. 요일 변수 추가

```python
평일_추가전['집계일자'] = pd.to_datetime(평일_추가전['집계일자'], format='%Y-%m-%d')
평일_추가전['요일'] = 평일_추가전['집계일자'].dt.day_name()
평일_추가전['요일'] = 평일_추가전['요일'].str.replace('Monday', '월')
평일_추가전['요일'] = 평일_추가전['요일'].str.replace('Tuesday', '화')
평일_추가전['요일'] = 평일_추가전['요일'].str.replace('Wednesday', '수')
평일_추가전['요일'] = 평일_추가전['요일'].str.replace('Thursday', '목')
평일_추가전['요일'] = 평일_추가전['요일'].str.replace('Friday', '금')
평일_추가전
```

```python
주말_추가전['집계일자'] = pd.to_datetime(주말_추가전['집계일자'], format='%Y-%m-%d')
주말_추가전['요일'] = 주말_추가전['집계일자'].dt.day_name()
주말_추가전['요일'] = 주말_추가전['요일'].str.replace('Saturday', '토')
주말_추가전['요일'] = 주말_추가전['요일'].str.replace('Sunday', '일')
주말_추가전
```

## 4-2. 차종별 교통량, 비율 변수 추가

```python
평일_합치기 = pd.merge(평일_추가전, 종별교통량, how='left', on=['콘존명', '집계시'])
평일_합치기['승용차비율'] = 평일_합치기['승용차']/평일_합치기['계']
평일_합치기['버스비율'] = 평일_합치기['버스']/평일_합치기['계']
평일_합치기['화물차비율'] = 평일_합치기['화물차']/평일_합치기['계']
평일_합치기 = 평일_합치기.drop(['본부', '노선', '방향'],axis=1)
평일_합치기
```

```python
주말_합치기 = pd.merge(주말_추가전, 종별교통량, how='left', on=['콘존명', '집계시'])
주말_합치기['승용차비율'] = 주말_합치기['승용차']/주말_합치기['계']
주말_합치기['버스비율'] = 주말_합치기['버스']/주말_합치기['계']
주말_합치기['화물차비율'] = 주말_합치기['화물차']/주말_합치기['계']
주말_합치기 = 주말_합치기.drop(['본부', '노선', '방향'],axis=1)
주말_합치기
```

## 4-3. 차로 개수 변수 추가

```python
평일_합치기 = pd.merge(평일_합치기, 수도권_차로, how="left", on=['VDS_ID'])
평일_합치기 = 평일_합치기.drop(['지점이정','콘존명_y'], axis=1)
평일_합치기 = 평일_합치기.rename(columns={'콘존명_x':'콘존명'})
평일_합치기
```

```python
주말_합치기 = pd.merge(주말_합치기, 수도권_차로, how="left", on=['VDS_ID'])
주말_합치기 = 주말_합치기.drop(['지점이정','콘존명_y'], axis=1)
주말_합치기 = 주말_합치기.rename(columns={'콘존명_x':'콘존명'})
주말_합치기
```

## 5. 컬럼 순서 변경

```python
평일_합치기 = 평일_합치기[['집계일자','요일','집계시','VDS_ID','콘존명','차로개수','교통량','평균통행속도','평균통행시간','거리',
                           '계','승용차','버스','화물차','승용차비율','버스비율','화물차비율']]
평일_합치기
```

```python
주말_합치기 = 주말_합치기[['집계일자','요일','집계시','VDS_ID','콘존명','차로개수','교통량','평균통행속도','평균통행시간','거리',
                           '계','승용차','버스','화물차','승용차비율','버스비율','화물차비율']]
주말_합치기
```

# 6. 빈셀 삭제

```python
평일_합치기 = 평일_합치기[평일_합치기['계'].notna()]
주말_합치기 = 주말_합치기[주말_합치기['계'].notna()]
```

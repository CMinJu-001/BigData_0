# 2. 지표 계산

### 변수 선정

```python
재차인원_승용차 = 1.7
재차인원_버스 = 16
재차인원_화물차 = 1.2

적용후_버스_재차인원_18명 = 18
적용후_버스_재차인원_20명 = 20
적용후_버스_재차인원_22명 = 22

적용후_승용차_감소율_5퍼 = 0.95
적용후_승용차_감소율_7_5퍼 = 0.925
적용후_승용차_감소율_10퍼 = 0.9

일반차로_속도변화2_5퍼감소 = 0.975
일반차로_속도변화5퍼감소 = 0.95
일반차로_속도변화7_5퍼감소 = 0.925
일반차로_속도변화10퍼감소 = 0.9

버스차로_속도변화 = 0.95 # 다른 차들보다 약간 느리게 설정
```

## 2-1. 외곽선_평일

### 1) 기존_수송인원 구하기

```python
외곽선_평일['기존_수송인원'] = (외곽선_평일['교통량']*외곽선_평일['승용차비율']*재차인원_승용차)+(외곽선_평일['교통량']*외곽선_평일['버스비율']*재차인원_버스)+(외곽선_평일['교통량']*외곽선_평일['화물차비율']*재차인원_화물차)
외곽선_평일
```

### 2) 기존_총통행시간 구하기

```python
외곽선_평일['기존_총통행시간'] = 외곽선_평일['평균통행시간']*외곽선_평일['기존_수송인원']
외곽선_평일
```

```python
외곽선_평일 = 외곽선_평일.drop(['승용차','버스','화물차','계'],axis=1)
외곽선_평일
```

### 3) 버스전용차로 적용 시

```python
외곽선_평일_적용후 = pd.DataFrame(columns=['집계일자','요일','집계시','VDS_ID','콘존명','차로개수',
                                           '일반차로 교통량','일반차로 통행속도','일반차로 통행시간','일반차로 수송인원',
                                           '버스차로 교통량','버스차로 통행속도','버스차로 통행시간','버스차로 수송인원'])
외곽선_평일_적용후
```

```python
list1 = [일반차로_속도변화2_5퍼감소,일반차로_속도변화5퍼감소,일반차로_속도변화7_5퍼감소,일반차로_속도변화10퍼감소]
list2 = [적용후_버스_재차인원_18명,적용후_버스_재차인원_20명,적용후_버스_재차인원_22명]
list3 = [적용후_승용차_감소율_5퍼,적용후_승용차_감소율_7_5퍼,적용후_승용차_감소율_10퍼]

for i in list1:
    for j,k in zip(list2,list3):
        외곽선_평일_적용후['집계일자'] = 외곽선_평일['집계일자']
        외곽선_평일_적용후['요일'] = 외곽선_평일['요일']
        외곽선_평일_적용후['집계시'] = 외곽선_평일['집계시']
        외곽선_평일_적용후['VDS_ID'] = 외곽선_평일['VDS_ID']
        외곽선_평일_적용후['콘존명'] = 외곽선_평일['콘존명']
        외곽선_평일_적용후['차로개수'] = 외곽선_평일['차로개수']
        외곽선_평일_적용후['버스차로 교통량'] = 외곽선_평일['교통량']*외곽선_평일['버스비율']
        외곽선_평일_적용후['일반차로 교통량'] = 외곽선_평일['교통량']-외곽선_평일_적용후['버스차로 교통량']
        외곽선_평일_적용후['일반차로 통행속도'] = 외곽선_평일['평균통행속도']*i #속도변화
        외곽선_평일_적용후['버스차로 통행속도'] = 외곽선_평일['평균통행속도']*버스차로_속도변화
        외곽선_평일_적용후['일반차로 통행시간'] = 외곽선_평일['거리']/외곽선_평일_적용후['일반차로 통행속도']*3600
        외곽선_평일_적용후['버스차로 통행시간'] = 외곽선_평일['거리']/외곽선_평일_적용후['버스차로 통행속도']*3600
        외곽선_평일_적용후['일반차로 수송인원'] = (외곽선_평일_적용후['일반차로 교통량']*외곽선_평일['승용차비율']*재차인원_승용차*k)+(외곽선_평일_적용후['일반차로 교통량']*외곽선_평일['화물차비율']*재차인원_화물차)
        외곽선_평일_적용후['버스차로 수송인원'] = 외곽선_평일_적용후['버스차로 교통량']*j #버스재차인원
        외곽선_평일_적용후[f'일속변_{i} 버재_{j} 승감_{k}'] = (외곽선_평일_적용후['일반차로 통행시간']*외곽선_평일_적용후['일반차로 수송인원'])+(외곽선_평일_적용후['버스차로 통행시간']*외곽선_평일_적용후['버스차로 수송인원'])
        외곽선_평일_적용후['기존_총통행시간'] = 외곽선_평일['기존_총통행시간']
        외곽선_평일_적용후 = 외곽선_평일_적용후[외곽선_평일_적용후['차로개수']>=3]
        외곽선_평일_적용후.to_csv(f'C:/python/3. 결과데이터/외곽선_평일/일속변_{i} 버재_{j} 승감_{k}.csv', encoding='EUC-KR', mode='w')
        외곽선_평일_적용후 = pd.DataFrame(columns=['집계일자','요일','집계시','VDS_ID','콘존명','차로개수',
                                                   '일반차로 교통량','일반차로 통행속도','일반차로 통행시간','일반차로 수송인원',
                                                   '버스차로 교통량','버스차로 통행속도','버스차로 통행시간','버스차로 수송인원'])
```

```python
list1 = [일반차로_속도변화2_5퍼감소,일반차로_속도변화5퍼감소,일반차로_속도변화7_5퍼감소,일반차로_속도변화10퍼감소]
list2 = [적용후_버스_재차인원_18명,적용후_버스_재차인원_20명,적용후_버스_재차인원_22명]
list3 = [적용후_승용차_감소율_5퍼,적용후_승용차_감소율_7_5퍼,적용후_승용차_감소율_10퍼]

for i in list1:
    for j,k in zip(list2,list3):
        외곽선_주말_적용후['집계일자'] = 외곽선_주말['집계일자']
        외곽선_주말_적용후['요일'] = 외곽선_주말['요일']
        외곽선_주말_적용후['집계시'] = 외곽선_주말['집계시']
        외곽선_주말_적용후['VDS_ID'] = 외곽선_주말['VDS_ID']
        외곽선_주말_적용후['콘존명'] = 외곽선_주말['콘존명']
        외곽선_주말_적용후['차로개수'] = 외곽선_주말['차로개수']
        외곽선_주말_적용후['버스차로 교통량'] = 외곽선_주말['교통량']*외곽선_주말['버스비율']
        외곽선_주말_적용후['일반차로 교통량'] = 외곽선_주말['교통량']-외곽선_주말_적용후['버스차로 교통량']
        외곽선_주말_적용후['일반차로 통행속도'] = 외곽선_주말['평균통행속도']*i #속도변화
        외곽선_주말_적용후['버스차로 통행속도'] = 외곽선_주말['평균통행속도']*버스차로_속도변화
        외곽선_주말_적용후['일반차로 통행시간'] = 외곽선_주말['거리']/외곽선_주말_적용후['일반차로 통행속도']*3600
        외곽선_주말_적용후['버스차로 통행시간'] = 외곽선_주말['거리']/외곽선_주말_적용후['버스차로 통행속도']*3600
        외곽선_주말_적용후['일반차로 수송인원'] = (외곽선_주말_적용후['일반차로 교통량']*외곽선_주말['승용차비율']*재차인원_승용차*k)+(외곽선_주말_적용후['일반차로 교통량']*외곽선_주말['화물차비율']*재차인원_화물차)
        외곽선_주말_적용후['버스차로 수송인원'] = 외곽선_주말_적용후['버스차로 교통량']*j #버스재차인원
        외곽선_주말_적용후[f'일속변_{i} 버재_{j} 승감_{k}'] = (외곽선_주말_적용후['일반차로 통행시간']*외곽선_주말_적용후['일반차로 수송인원'])+(외곽선_주말_적용후['버스차로 통행시간']*외곽선_주말_적용후['버스차로 수송인원'])
        외곽선_주말_적용후['기존_총통행시간'] = 외곽선_주말['기존_총통행시간']
        외곽선_주말_적용후 = 외곽선_주말_적용후[외곽선_주말_적용후['차로개수']>=3]
        외곽선_주말_적용후.to_csv(f'C:/python/3. 결과데이터/외곽선_주말/일속변_{i} 버재_{j} 승감_{k}.csv', encoding='EUC-KR', mode='w') # mode="w" 덮어쓰기 옵션
        외곽선_주말_적용후 = pd.DataFrame(columns=['집계일자','요일','집계시','VDS_ID','콘존명','차로개수',
                                                   '일반차로 교통량','일반차로 통행속도','일반차로 통행시간','일반차로 수송인원',
                                                   '버스차로 교통량','버스차로 통행속도','버스차로 통행시간','버스차로 수송인원'])
```
